```python
def json_to_dataframe(response):
    return pd.DataFrame(response.json()[1:], columns = response.json()[0])
```

```python
url = "https://api.census.gov/data/2010/dec/sf1?get=H001001,P001001,P013001,H012001,H003001&for=zip%20code%20tabulation%20area%20(or%20part):*&in=state:06&key={0}".format(api_key)
r = requests.request("GET", url)
```

```python
api_df = json_to_dataframe(r)
```

```python
api_df['Housing Units']= pd.to_numeric(api_df['H001001'])
api_df['Total Population']= pd.to_numeric(api_df['P001001'])
api_df['Median Age']= pd.to_numeric(api_df['P013001'])
api_df['Avg household size']= pd.to_numeric(api_df['H012001'])
api_df['Occupancy']= pd.to_numeric(api_df['H003001'])
api_df['zip code'] = pd.to_numeric(api_df['zip code tabulation area (or part)'])
api_df = api_df.drop(['state','zip code tabulation area (or part)','H001001','P001001','P013001','H012001','H003001'], axis=1)
```
